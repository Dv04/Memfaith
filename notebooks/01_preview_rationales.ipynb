# Preview Rationales
Small sanity check for EF prompts.

```python
from src.data_loading import load_dataset
from src.rationale_model import RationaleLM

examples = load_dataset("fever", "train")[:5]
lm = RationaleLM("openai-community/gpt2", device="cpu", max_new_tokens=64)
outputs = lm.generate_answer_and_rationale(examples)
for out in outputs:
    print(out.example.example_id, out.model_answer)
    print(out.rationale)
    print('-' * 40)
```
