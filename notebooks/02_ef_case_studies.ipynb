# EF Case Studies
Inspect target vs control flips.

```python
import json
from pathlib import Path

log_path = Path('experiments/logs/fever_gpt2xl_rome.jsonl')
records = [json.loads(line) for line in log_path.read_text().splitlines() if line.strip()]
len(records)
```

```python
def show_case(record):
    print('Example ID:', record['example_id'])
    print('Claim/Question:', record['input_text'])
    print('Original answer:', record['orig_answer'])
    print('Target triple:', record['target_triple'])
    print('Control triple:', record['control_triple'])
    print('Post target answer:', record['post_edit_answer_target'])
    print('Post control answer:', record['post_edit_answer_control'])
    print('---')

for rec in records[:3]:
    show_case(rec)
```
